```ruby
require 'json'

class Node
  
  attr_accessor :children, :name, :audience_size
  
  def initialize(name,audience=nil)
    @name = name
    @audience_size = audience
    @children = {}
  end
  
  def add(node)
    @children[node.name] = node

  end
  
  def [](name)
    @children[name]
  end
  
  def size
    sum_audience(self)
  end
  
  def is_leaf?
    @children.empty?
  end
  
  def find(node_name)
    do_find(node_name,self)
  end
  
  def print
    puts name
    do_print(self)
  end
  
  def print_leafs
    do_print_leafs(self)
  end
  
  def print_proportions
    total = self.size
    puts "#{self.name}: #{self.proportion(total)}"
    do_print_proportions(self,total)
  end
  
  def each_leaf(&block) 
    block.call
  end
  
  def proportion(total=56274743664)
    "#{(self.size/(total + 0.0))*100}%"
  end

  private

  def do_print_leafs(node)
    if node.is_leaf?
      puts node.name
    else
      node.children.values.each do |child|
        do_print(child)
      end
    end
  end
  
  def do_print(node)
    if node.is_leaf?
      puts node.name
      return false
    else
      node.children.values.each do |child|
        if do_print(child)
          puts child.name
        end
      end
    end
  end
  
  def do_print_proportions(node,size)
    if node.is_leaf?
      puts "#{node.name}: #{node.proportion(size)}"
      return false
    else
      node.children.values.each do |child|
        if do_print_proportions(child,size)
          puts "#{child.name}: #{child.proportion(size)}"
        end
      end
    end
  end

  # Definetly not the best way, but it works...
  def do_find(node_name,node)
    if node.name == node_name
      @found_node = node
    else
      node.children.values.each do |child|
        do_find(node_name,child)
      end
    end
    return @found_node
  end
  
  def level(node_name,node)
    if node.name == node_name
      @found_node = node
    else
      node.children.values.each do |child|
        do_find(node_name,child)
      end
    end
    return @found_node
  end
  
  #def sum_audience(node)
  #  total = 0
  #  if node.is_leaf?
  #    return node.audience_size
  #  else
  #    node.children.values.each do |child|
  #      total += sum_audience(child)
  #    end
  #  end
  #  return total
  #end
  
    def sum_audience(node)
      total = 0
      if node.is_leaf?
        return node.audience_size
      else
        node.children.values.each do |child|
          if child.is_leaf?
            total += child.audience_size
          else
            total += sum_audience(child)
          end
        end
      end
      return total
    end
end

n = Node.new("Root")
n2 = Node.new("Left")
n3 = Node.new("Right")
n4 = Node.new("Left Left")
n5 = Node.new("Left Right")
n6 = Node.new("Left Left Left")
n7 = Node.new("Right Left Right")

n4.add(n6)
n5.add(n7)
n2.add(n4)
n3.add(n5)
n.add(n2)
n.add(n3)

n.print
n.find("Left Right")
```

```ruby
require 'json'

tree = JSON.parse(File.read("interests_tree.json"))

def parseTree(tree)
  
  n = tree.shift
  root = Node.new(n["name"])
  
  tree.each do |node|
    if node["audience_size"] == nil
      leaf = Node.new(node["name"])
    else
      leaf = Node.new(node["name"],node["audience_size"])
    end
    # Again!, not the best, but it works...
    root.find(node["path"].last).add(leaf)
  end
  root
end

node_tree = parseTree(tree)
puts node_tree.size
node_tree.print_proportions
```

```ruby
require 'faraday'
require 'json'

conn = Faraday.new(url: "https://graph.facebook.com")

params = {
	"age_max": 65,
	"user_device": [],
	"excluded_publisher_list_ids": [],
	"flexible_spec": [{
		"interests": [{
			"id": "6003840140052",
			"name": "Agriculture"
		}]
	}],
	"geo_locations": {
		"location_types": ["home", "recent"],
		"countries": ["MX"]
	},
	"facebook_positions": ["feed"],
	"age_min": 13,
	"excluded_publisher_categories": [],
	"excluded_user_device": [],
	"wireless_carrier": [],
	"device_platforms": ["mobile", "desktop"],
	"user_os": [],
	"instagram_positions": [],
	"publisher_platforms": ["facebook"],
	"audience_network_positions": [],
	"messenger_positions": []
}

def get_audience_size(id,name,params,conn)
  puts "Getting info for... #{name}"
  params[:flexible_spec][0][:interests] = [{id: id, name: name}]
  res = conn.get "/v2.11/act_70665620/delivery_estimate", {
  access_token: "EAABsbCS1iHgBAJcRLtZBhM2dIZCoz9ZAQphWShlrBubZBBWIOZC0mPDNbfOZCpvVdNu1OKyFgwKaZAem8m7IcedPG36TB4H7FTgmJKFzGvjZA4rUb5KmsbNw0m92GJaSWF2miJZBY3bfgQsZAhZCFx6GSKrIzzcYMx5HMAZD",
  optimization_goal: "EVENT_RESPONSES",
  targeting_spec: params.to_json
  }
  
  json = JSON.parse(res.body)
  sleep(1)
  return json["data"][0]["estimate_mau"]
end

tree = JSON.parse(File.read("interests_tree.json"))

def write_custom_interests(tree)
  File.open("custom_interests.json","w") do |file|
    file.puts tree.to_json
  end
end

def parseTree(tree,params,conn)
  
  n = tree.shift
  root = Node.new(n["name"])
  
  tree.each do |node|    
    if node["audience_size"] == nil
      leaf = Node.new(node["name"])
    else
      audience = get_audience_size(node["id"],node["name"],params,conn)
      leaf = Node.new(node["name"], audience)
      node["audience_size"] = audience
    end
    # Again!, not the best, but it works...
    root.find(node["path"].last).add(leaf)
  end
  write_custom_interests(tree)
  root
end

node_filtered_tree = parseTree(tree,params,conn)
```
